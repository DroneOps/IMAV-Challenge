```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the file path to the CSV file containing the data
file_path = 'ErrorData/error_data_test.csv'  # Replace with your actual file path

# Read the data from the file
data = np.loadtxt(file_path, delimiter=',', skiprows=1)

# Sort the data by the first column (time steps) to ensure correct plotting
data = data[data[:, 0].argsort()]

# Extract time steps, Y errors, and Z errors
time_steps = data[0:26533, 0]
y_errors = data[0:26533, 1]
z_errors = data[0:26533, 2]

seconds = time_steps/1  # Convert time steps to seconds by normalizing to the first time step

# Create a figure and axis for plotting
plt.figure(figsize=(10, 6))

# Plot Y errors
plt.plot(seconds, y_errors, label='Y Error', color='blue')

# Plot Z errors
plt.plot(seconds, z_errors, label='Z Error', color='orange')

# Add labels and title
plt.xlabel('Time (seconds)')
plt.ylabel('Error Values')
plt.title('Y and Z Errors Over Time')

# Add legend
plt.legend()

# Show grid
plt.grid()

# Display the plot
plt.show()
```
